```python
# importing important libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
## loading the dataste
data=load_boston()
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['MEDV']=data.target
```

```python
df.head()
```

```python
## splitting data into Independent and Dependent variable.
X=df.drop('MEDV',axis=1)
y=df['MEDV']
```

```python
X.head()
```

```python
y.head()
```

```python
## Splitting data into TRaining and Testing Dataset.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
cor=x_train.corr()
cor 
```

### Pearson Correlation

```python
plt.figure(figsize=(12,8))
cor=x_train.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
## Creating the function for find out the correlation features.
def get_correlation(data,threshold):
    corr_col=set()
    cormat=data.corr()
    for i in range(len(cormat.columns)):
        for j in range(i):
            if abs(cormat.iloc[i,j])>threshold:
                colname=cormat.columns[i]
                corr_col.add(colname)
    return corr_col
```

```python
corr_features=get_correlation(x_train,0.7)
print(len(corr_features))
```

```python
corr_features
```

#### Dropping correlated features.

```python
x_train_new=x_train.drop(corr_features,axis=1)
x_test_new=x_test.drop(corr_features,axis=1)
```

```python
x_train_new
```

```python
x_test_new
```
