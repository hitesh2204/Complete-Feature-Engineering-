## FeatureSelectionTechnique.

```python
import pandas as pd
df=pd.read_csv("E://Practise//FeatureEngineering//mobile_dataset.csv")
df.head()
```

###  Univariate Selection.

```python

x=df.iloc[:,:-1]
y=df['price_range']
x.head()
```

```python
y.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
##Apply Select KBest algorithm.
order_rank_feature=SelectKBest(score_func=chi2,k=20)
order_rank_feature.fit(x,y)
```

```python
order_rank_feature.scores_
```

```python
dfscore=pd.DataFrame(order_rank_feature.scores_,columns=['Score'])
dfcolumns=pd.DataFrame(x.columns)
```

```python
feature_rank=pd.concat([dfcolumns,dfscore],axis=1)
feature_rank.columns=['Feature','Score']
feature_rank
```

```python
feature_rank.nlargest(10,'Score')
```

#### Feature Importance.

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model=ExtraTreesClassifier()
model.fit(x,y)
```

```python
ranked_feature=pd.Series(model.feature_importances_,index=x.columns)
ranked_feature.nlargest(10).plot(kind='barh')
plot.show()
```
